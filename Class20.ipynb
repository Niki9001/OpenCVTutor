# Image Dilation

## What is image dilation?

Image Dilation is an image processing technique used to increase the size of foreground objects in an image. Think of it as drawing circles around a pattern with a pen to make the whole pattern look bigger. The "pen" used here corresponds to a shape in image processing called a "structuring element" or "kernel", which slides across the image. When any part of the kernel covers a foreground pixel (i.e., a point with a value of 1, or the white part), the center pixel point covered by the kernel is set to the foreground value.

This operation causes the edges of objects in the image to expand outward, which can be used to fill small holes in an object, connect neighboring objects, or increase the size of an object. Image expansion has the opposite effect of erosion; instead of decreasing, it increases the area of a foreground object, which is used in image processing to strengthen or thicken the outline of an object.

Image dilation is the inverse operation of image erosion

## Syntax

## Get greyscale image

```python
import cv2
path = "H:\\openCV\\test.jpg"
# be careful of path and \\
image = cv2.imread(path,flags=0)
# check if the image load successfully or not 
if image is None:
    print("Fail to load")
else:
    cv2.imshow('image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import numpy as np
kernel = np.ones((5,5))
dilate = cv2.dilate(image,kernel)
cv2.imshow('dilate', dilate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

We can see the changes of this image

```python
kernel = np.ones((5,5))
dilate5 = cv2.dilate(image,kernel,iterations=5 )
cv2.imshow('dilate5', dilate5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

After 5 times processing, the result changed a lot
