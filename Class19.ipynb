# Image Erosion

## What is image erosion?

Image Erosion (Erosion) is an image processing technique that is used to reduce the size of foreground objects in an image. Imagine that you are scrubbing a pattern with a small brush, with the goal of removing only the edges, making the whole pattern look smaller. In image processing, this brush corresponds to a shape called a "structuring element" or "kernel," which slides over the image and decides whether to reduce the value of the current pixel (usually to a background value, such as black) based on how closely the kernel overlaps the image.

The effect of this operation is that the details of smaller objects or edges in the image are "etched away", leaving cleaner shapes. This is very useful for removing small noises, disconnecting neighboring objects, and so on.

## How does it work

In OpenCV, the process of image erosion for binary images, where 0 represents black (background) and 1 represents white (foreground), involves applying an operation that focuses on reducing the white regions. Erosion typically requires two parameters: the binary image (original image) and the convolution kernel (also known as the structuring element).

```python
Image(filename="H:\\openCV\\erosion2.png")
```

For instance, consider a scenario where the convolution kernel, represented by red and yellow parts, moves over the image. When the kernel is entirely surrounded by white pixels, the central pixel's value remains white (1) after the operation. This is because the entire area covered by the kernel matches the foreground condition. Conversely, in the case illustrated by the right side of an example, where the convolution kernel encounters black pixels (background), the outcome for that central position after processing turns to 0, which means it becomes black.

This process is effectively used to shrink the white regions, thereby reducing the size of foreground objects in the binary image. The kernel slides over the image, and for each position, if all the pixels under the kernel are white, the central pixel remains white. If any pixel under the kernel is black, the central pixel is set to black. This operation is helpful for removing small imperfections or separating objects that are close to each other in binary images.

## syntax

image = cv2.erode(source, kernel, iterations)

kernel usually is a numpy array

iterations mean how many times erosion we want to. The default value is 1

## Get greyscale image

```python
import cv2
path = "H:\\openCV\\test.jpg"
# be careful of path and \\
image = cv2.imread(path,flags=0)
# check if the image load successfully or not 
if image is None:
    print("Fail to load")
else:
    cv2.imshow('image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import numpy as np
kernel = np.ones((5,5))
erosion = cv2.erode(image,kernel)
cv2.imshow('erosion', erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

We can see the result is slightly different, that means our code works, however when we use erosion in real time, usually we need to do more work in advance to ensure the result

```python
kernel = np.ones((5,5))
erosion9 = cv2.erode(image,kernel,iterations=5)
cv2.imshow('erosion9', erosion9)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

If we set the iterations value is higher, we can see the difference of the result
