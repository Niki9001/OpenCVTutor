# Gaussian filter

## What is Gaussian filter

Gaussian filtering is a method used in image processing to reduce noise and detail from images. In plain English, imagine you have a photograph where some parts look grainy or speckled; Gaussian filtering is like smoothing out those rough spots to make the image clearer and more pleasing to the eye.

To do this, Gaussian filtering uses a special kind of mathematical formula that calculates the average color of a pixel based on its neighbors, giving more weight to those that are closer. This process is similar to blurring, but it's done in a very controlled way that preserves the overall shapes and edges in the picture. It's named "Gaussian" because the weights are determined by the Gaussian function, which is shaped like a bell curve.

In practical terms, when you apply Gaussian filtering to an image, it's like using a soft brush to gently blur away imperfections, helping important details to stand out more clearly while reducing distracting noise.

## Get colorful image

```python
import cv2
path = "H:\\openCV\\test.jpg"
# be careful of path and \\
imageColor = cv2.imread(path)
# check if the image load successfully or not 
if imageColor is None:
    print("Fail to load")
else:
    cv2.imshow('imageColor', imageColor)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
core3Gaussian = cv2.GaussianBlur(imageColor,(3,3),0)
core7Gaussian = cv2.GaussianBlur(imageColor,(7,7),0)

cv2.imshow('core=7 Gaussian', core7Gaussian)
cv2.imshow('core=3 Gaussian', core3Gaussian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

We can see the difference between core=3 and core=7
